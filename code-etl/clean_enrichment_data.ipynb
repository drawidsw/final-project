```python
import os
import pandas as pd
```

```python
# Utility function to return a "cleaned school" name.
def clean_school_name(s):
    
    arr = s.split()
    
    # If the last word is 'School', drop it.
    if arr[len(arr)-1].lower() == 'school':
        arr.pop()
        
    # Convert all words to lowercase.
    for i in range(0, len(arr)):
        arr[i] = arr[i].lower()
        
    return ' '.join(arr)
```

```python
# Load the school ratings data.
schools_df = pd.read_csv(os.path.join('..', 'csv_data', 'mcps_school_ratings.csv'))
schools_df.head()
```

```python
# Iterate over all schools.
for index, row in schools_df.iterrows():
    
    clean_sn = clean_school_name(row['School'])
    schools_df.loc[index, 'School'] = clean_sn
    
schools_df = schools_df[['School', 'Rating', 'Type']]

schools_df
```

```python
schools_df.to_csv(os.path.join('..', 'csv_data', 'mcps_school_ratings_clean.csv'), index=False)
```

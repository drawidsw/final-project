```python
import os
import pandas as pd
from datetime import datetime
```

```python
houses_df = pd.read_csv(os.path.join('..', 'csv_data', 'house_info_school_assignments_dropna.csv'))
houses_df
```

```python
# keep only columns we need for the preliminary machine learning.
# Keep enrichment keys (zipcode and all type of schools)
houses_df = houses_df[
                        [ 'price',
                          'dateSold',
                          'bathrooms',
                          'bedrooms',
                          'livingArea',
                          'homeType',
                          'lotAreaValue',
                          'lotAreaUnit',
                          'zipcode',
                          'elem',
                          'middle',
                          'high',
                          'Year built'
                        ]
                     ]

houses_df
```

```python
# Drop all rows with even a single null value.
houses_df = houses_df.dropna()
houses_df
```

```python
# Convert all lot areas to acres.
for index, row in houses_df.iterrows():
    
    if row['lotAreaUnit'] == 'acres':
        val = houses_df.loc[index, 'lotAreaValue'] * 43560
        houses_df.loc[index, 'lotAreaValue'] = val

# Drop the lotAreaUnit column.
houses_df = houses_df.drop(columns=['lotAreaUnit'])

houses_df
```

```python
# Drop rows with housetypes apartments, multi-family and lot. There are too few of them.
houses_df.drop(houses_df[houses_df.homeType == 'APARTMENT'].index, inplace=True)
houses_df.drop(houses_df[houses_df.homeType == 'LOT'].index, inplace=True)
houses_df.drop(houses_df[houses_df.homeType == 'MULTI_FAMILY'].index, inplace=True)
houses_df
```

```python
# Convert dateSold to dates.
houses_df['dateSold_month'] = houses_df['dateSold'].apply(lambda x: datetime.fromtimestamp(x/1000).month)
houses_df['dateSold_year'] = houses_df['dateSold'].apply(lambda x: datetime.fromtimestamp(x/1000).year)

for index, row in houses_df.iterrows():
    
    if row['dateSold_year'] == 2020:
        if row['dateSold_month'] < 4:
            houses_df.loc[index, 'dateSold'] = 'Q1_2020'
        elif row['dateSold_month'] < 7:
            houses_df.loc[index, 'dateSold'] = 'Q2_2020'
        elif row['dateSold_month'] < 10:
            houses_df.loc[index, 'dateSold'] = 'Q3_2020'
        else:
            houses_df.loc[index, 'dateSold'] = 'Q4_2020'
            
    if row['dateSold_year'] == 2021:
        if row['dateSold_month'] < 4:
            houses_df.loc[index, 'dateSold'] = 'Q1_2021'
        elif row['dateSold_month'] < 7:
            houses_df.loc[index, 'dateSold'] = 'Q2_2021'
        elif row['dateSold_month'] < 10:
            houses_df.loc[index, 'dateSold'] = 'Q3_2021'
        else:
            houses_df.loc[index, 'dateSold'] = 'Q4_2021'

houses_df = houses_df.drop(columns=['dateSold_month', 'dateSold_year'])
houses_df
```

```python
# Convert year built.
from datetime import date

houses_df['AgeYears'] = houses_df['Year built'].apply(lambda x: date.today().year-x)
houses_df = houses_df.drop(columns=['Year built'])
houses_df
```

```python
houses_df.to_csv(os.path.join('..', 'csv_data', 'house_info_school_assignments_dropna_transform.csv'), index=False)
```
